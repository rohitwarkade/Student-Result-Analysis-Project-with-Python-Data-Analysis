```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
# HEATMAP
```

```python
plt.figure(figsize=(5,5)) # size 
sns.heatmap(gb)
plt.title(" RELATIONSHIP BETWEEN PARENTS'S EDUCATION AND STUDENT'S SCORE")
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb, annot=True)
plt.show()
```

```python
# FROM THE ABOVE CHART WE HAVE CONCLUDED THAT THE EDUCATION OF  THE PARENTS HAVE A GOOD IMPACT
```
